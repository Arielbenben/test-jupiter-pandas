```python
url_Israel_interest_tares = './BR.csv'
Israel_interest_tares_csv = pd.read_csv(url_Israel_interest_tares)
Israel_interest_tares = pd.DataFrame({
    'Year': [year[0:4] for year in Israel_interest_tares_csv['TIME_PERIOD']],
    'Month': [month[5:7] for month in Israel_interest_tares_csv['TIME_PERIOD']],
    'Current_Rate': [rate for rate in Israel_interest_tares_csv['OBS_VALUE']],
})
Israel_interest_tares['Change_From_Last_Month'] = Israel_interest_tares['Current_Rate'].diff()
Israel_interest_tares['Change_From_Last_Month'] = Israel_interest_tares['Change_From_Last_Month'].fillna(0)


url_Usa_interest_tares = './FEDFUNDS.csv'
Usa_interest_tares_csv = pd.read_csv(url_Usa_interest_tares)
Usa_interest_tares = pd.DataFrame({
    'Year': [year[0:4] for year in Usa_interest_tares_csv['DATE']],
    'Month': [month[5:7] for month in Usa_interest_tares_csv['DATE']],
    'Current_Rate': [rate for rate in Usa_interest_tares_csv['FEDFUNDS']],
})
Usa_interest_tares['Change_From_Last_Month'] = Usa_interest_tares['Current_Rate'].diff()
Usa_interest_tares['Change_From_Last_Month'] = Usa_interest_tares['Change_From_Last_Month'].fillna(0)


url_USD_to_ILS_exchange_rates = './additional_rates_19930101_20230101_en_USD.csv'
USD_to_ILS_exchange_rates_csv = pd.read_csv(url_USD_to_ILS_exchange_rates)
USD_to_ILS_exchange_rates = pd.DataFrame({
    'Year': [year[6:] for year in USD_to_ILS_exchange_rates_csv['Date    ']],
    'Month': [month[3:5] for month in USD_to_ILS_exchange_rates_csv['Date    ']],
    'Current_Rate': [rate for rate in USD_to_ILS_exchange_rates_csv['USD']],
})
USD_to_ILS_exchange_rates['Change_From_Last_Month'] = USD_to_ILS_exchange_rates['Current_Rate'].diff()
USD_to_ILS_exchange_rates['Change_From_Last_Month'] = USD_to_ILS_exchange_rates['Change_From_Last_Month'].fillna(0)
```

```python
# B - 1
merge_df = pd.merge(Usa_interest_tares, Israel_interest_tares, on=['Year', 'Month'])
```

```python
# B - 2
merge_df['Rate_Difference'] = merge_df['Current_Rate_x'] - merge_df['Current_Rate_y']
merge_df
```

```python
# B - 3
merge_df['Rate_Comparison'] = merge_df.apply(
    lambda row: 'IH' if row['Current_Rate_y'] >= row['Current_Rate_x'] * 1.01 else 
                ('UH' if row['Current_Rate_x'] >= row['Current_Rate_y'] * 1.01 else 'N'), 
    axis=1
)
merge_df
```

```python
# B - 4
merge_all_df = pd.merge(merge_df, USD_to_ILS_exchange_rates, on=['Year', 'Month'])
merge_all_df
```

```python
# B - 5
correlation = merge_all_df['Rate_Difference'].corr(merge_all_df['Current_Rate'])
correlation

# You can see When the correlation is positive what means that when the difference between
# interest rates between the US and Israel increases
# (i.e., interest rates in the US rise faster than in Israel), the exchange rate
# (the value of the shekel against the dollar) tends to increase.
```
